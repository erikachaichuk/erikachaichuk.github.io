```python
import pandas as pd
from pandas import read_csv
import altair as alt
```

```python
prices_df = read_csv('https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_prices.csv')
prices_df = prices_df.drop(columns=['indicator_box', 'item_id_raw'])

prices_df.sample(5)
```

```python
# Sort and clean the dataset

items_df = pd.read_csv('https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_item_clean.csv')
items_df
```

```python
# Filter

sorted_items_df = items_df.sort_values(by='date_quote_s', ascending=True)
sorted_items_df = sorted_items_df[(sorted_items_df['date_quote_e'] >= 202100)]
sorted_items_df.head(30)
```

```python
sorted_items_df.to_csv('items.csv')
```

```python
# Filter for 3 items only

filtered_df = prices_df[prices_df['item_id'].isin([520216, 520209, 520213])]
filtered_df
```

```python
# Calculating mean and median

filtered_mean_df = filtered_df.groupby(['item_id', 'quote_date'])['price'].agg(['mean', 'median']).reset_index()
filtered_mean_df = filtered_mean_df.rename(columns={'mean': 'mean_price', 'median': 'median_price'})
filtered_mean_df
```

```python
# Adding description

final_df = pd.merge(filtered_mean_df, sorted_items_df, on='item_id')
final_df
```

```python
# Adding some characteristics

region_df = pd.read_csv("https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_region.csv")
region_df
```

```python
region_df.to_csv('regions.csv')
```

```python
import json

final_df['quote_date'] = final_df['quote_date'].dt.strftime('%Y-%m-%dT%H:%M:%S')
with open('personal_care_products.json', 'w') as f:
    json.dump(final_df.to_dict(orient='records'), f, indent=4)
```

Second graph


```python
weights_df = pd.read_csv("https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_itemWeights.csv")
weights_df = weights_df.drop(columns=['dup'])
weights_df['quote_date'] = pd.to_datetime(weights_df['quote_date'], format='%Y%m')
```

```python
df = prices_df.groupby(['region', 'quote_date', 'item_id'])['price'].mean().reset_index()

df
```

```python
df['quote_date'] = pd.to_datetime(df['quote_date'], format='%Y%m')

#Coping the data frame and adding last year to calculate inflation
df_new = df[['region', 'item_id', 'quote_date', 'price']].copy()
df_new['quote_date'] = df_new['quote_date'] + pd.DateOffset(years=1)
df_new = df_new.rename(columns={'price': 'last_year_price'})

df = pd.merge(df, df_new, on=['region', 'item_id', 'quote_date'], how='left')

df
```

```python
df['inflation'] = (df['price'] - df['last_year_price']) / df['last_year_price']

df
```

```python
# Merging using 'last year price' and 'price'

df_infl = df.drop(columns = ['last_year_price', 'price'])
df_infl = pd.merge(df_infl, weights_df, on=['item_id', 'quote_date'])

df_infl
```

```python
df_infl = df_infl.groupby(['region', 'quote_date'])['inflation'].mean().reset_index()

df_infl
```

```python
alt.Chart(df_infl).mark_line().encode(
    color=alt.Color('region:N'),
    x=alt.X('quote_date:T', title='', axis=alt.Axis(labels=True), scale=alt.Scale(domain=['1995-01-01', '2022-01-01'])),
    y=alt.Y('inflation', title='', scale=alt.Scale(domain=[-0.2, 0.5]))
).properties(
    title = {
        'text': "Inflation",
        'subtitle': ["Year on year inflation for each region"]
    },
    width=500)
```

```python
df_infl_new = df_infl[df_infl['inflation'] <= 0.5]

alt.Chart(df_infl_new).mark_line().encode(
    color=alt.Color('region:N'),
    x=alt.X('quote_date:T', title='', axis=alt.Axis(labels=True), scale=alt.Scale(domain=['1995-01-01', '2022-01-01'])),
    y=alt.Y('inflation', title='', scale=alt.Scale(domain=[-0.2, 0.5]))
).properties(
    title = {
        'text': "Inflation",
        'subtitle': ["Year on year inflation for each region"]
    },
    width=500)
```

```python
with open('region_inflation.json', 'w') as f:
    json.dump(df_infl_new.to_dict(orient='records'), f, indent=4)
```
